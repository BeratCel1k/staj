```python
from urllib import request
from bs4 import BeautifulSoup
import requests
from urllib.request import Request, urlopen

comment_link = []
liste = []
sayfa = 1

while sayfa <= 2:
    site = "https://www.hepsiburada.com/akilli-saatler-c-60003676?siralama=yorumsayisi&sayfa=" + str(sayfa)
    hdr = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"}
    req = requests.get(site, headers=hdr)
    soup = BeautifulSoup(req.content, "lxml")
    items = soup.find_all("div", attrs={"class": "productListContent-pXUkO4iHa51o_17CBibU"})

    for i in items:
        item = i.find_all("li", attrs={"type": "comfort"})
        for it in item:
            link_tail = it.a.get("href")
            link_head = "https://www.hepsiburada.com"
            a = 1
            while a <= 10:
                filter = 1
                while filter <= 5:
                    link = f"{link_head}{link_tail}-yorumlari?sayfa={a}&filtre={filter}"
                    site = link
                    hdr = {
                        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
                    req = requests.get(site, headers=hdr)
                    soup = BeautifulSoup(req.content, "lxml")
                    items = soup.find_all("div", attrs={"class": "paginationContentHolder"})
                    for i in items:
                        item = i.find_all("div", attrs={"itemprop": "review"})
                        for item_comment in item:
                            try:
                                comment = item_comment.find("span", attrs={"itemprop": "description"}).text
                            except:
                                print("NULL")
                            print(f"sayfa: {a}, filter: {filter}, comment: {comment}")
                            print("---")
                            liste.append([filter, comment])
                    filter += 1
                a += 1
            print("--------------------DiÄŸer Ã¼rÃ¼ne geÃ§iÅŸ---------------")
    sayfa += 1
    print("----------------------------------DiÄŸer sayfaya geÃ§iÅŸ-----------------------------------")
```
